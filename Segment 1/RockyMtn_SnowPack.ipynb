```python
import pandas as pd
```

```python
swe_df = pd.read_csv('../FP/colorado_headwaters.csv')
swe_df.head()
```

```python
def remove_columns(dataset,cols):
    return dataset.drop(cols, axis=1)
```

```python
swe_df = remove_columns(swe_df,["Min", "10%", "30%", "70%", "90%", "Max", "Median ('91-'20)", "Median (POR)", "Median Peak SWE"])
swe_df.head()
```

```python
import numpy as np
```

```python
swe_df = remove_columns(swe_df,["1986", "1987", "1988", "1989", "1990", "1991", "1992", "1993", "1994", "1995", "1996", "1997", "1998", "1999", "2000", "2001", "2023"])
swe_df.head()
```

```python
swe2_df = swe_df.set_index(["date"])
swe_reformatted_df = (
    swe2_df.stack([0])
    .rename("Value")
    .reset_index()
)
swe_reformatted_df.head(20)
```

```python
swe_reformatted_df = swe_reformatted_df.rename({"level_1":"Year"}, axis=1)
swe_reformatted_df.head(5)
```

```python
print (swe_reformatted_df.dtypes)
```

```python
from datetime import date
```

```python
swe_reformatted_df['DATE'] = pd.to_datetime(swe_reformatted_df.Year.astype(str) + '/' + swe_reformatted_df.date.astype(str))
swe_reformatted_df.head(30)
```

```python
swe_reformatted_df = remove_columns(swe_reformatted_df,["date", "Year"])
swe_reformatted_df.head(30)
```

```python
swe_reformatted_df.columns
```

```python
swe_reformatted_df = swe_reformatted_df[["DATE", "Value"]]
swe_reformatted_df
```

```python
swe_reformatted_df['year'] = pd.DatetimeIndex(swe_reformatted_df['DATE']).year
swe_reformatted_df['month'] = pd.DatetimeIndex(swe_reformatted_df['DATE']).month
swe_reformatted_df
```

```python
swe_reformatted_df = swe_reformatted_df.groupby([swe_reformatted_df['DATE'].dt.year, swe_reformatted_df['DATE'].dt.month]).mean('Value')
swe_reformatted_df
```

```python
swe_reformatted_df.dtypes
```

```python
swe_reformatted_df['year'] = swe_reformatted_df['year'].astype(int)
swe_reformatted_df['month'] = swe_reformatted_df['month'].astype(int)
swe_reformatted_df.dtypes
```

```python
swe_reformatted_df
```

```python
swe_reformatted_df = swe_reformatted_df.rename({"Value":"AvgSWE"}, axis=1)
swe_reformatted_df
```

```python
swe_reformatted_df.to_csv("colorado_headwaters_reformatted.csv", mode='w', header=True, index=False)
```
