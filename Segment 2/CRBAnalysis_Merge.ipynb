```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
# Import the KMeans module and matplotlib
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
drought_df = pd.read_csv('Segment 1/Final_Colorado_drought_data.csv')
drought_df.head()
```

```python
swe_df = pd.read_csv('Segment 1/colorado_headwaters_reformatted.csv')
swe_df.head()
```

```python
lm_df = pd.read_csv('Segment 1/lake_mead_new.csv')
lm_df.head()
```

```python
lp_df = pd.read_csv('Segment 1/lake_powell_new.csv')
lp_df.head()
```

```python
merged1_df = pd.merge(lp_df, lm_df, on=['year', 'month'])
```

```python
merged1_df.head()
```

```python
merged2_df = pd.merge(merged1_df, drought_df, on=['year', 'month'])
merged2_df.head()
```

```python
merged3_df = pd.merge(merged2_df, swe_df, on=['year', 'month'])
merged3_df.head()
```

```python
merged3_df = merged3_df.drop(["month_name_x", "month_name_y"],axis=1)
merged3_df.head()
```

```python
import os
```

```python
merged3_df = merged3_df.rename({"storage_x":"Powell_Storage", "storage_y":"Mead_Storage"}, axis=1)
merged3_df.head()
```

```python
merged3_df.dtypes
```

```python
merged3_df.loc[merged3_df["D_Combined"]<50.0,"D_Outcome"]=0
merged3_df.loc[merged3_df["D_Combined"]>=50.0,"D_Outcome"]=1
merged3_df.head(5)
```

```python
merged3_df.dtypes
```

```python
merged3_df["D_Outcome"]=merged3_df["D_Outcome"].astype(int)
merged3_df.dtypes
```

```python
merged3_df.to_csv("CRB_merged.csv", mode='w', header=True, index=False)
```
